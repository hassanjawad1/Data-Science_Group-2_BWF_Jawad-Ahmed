```python
FILE_NAME = 'My+Files.txt' 
import os

def load_data():
    students = {}
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, 'r') as file:
            for line in file:
                student_id, student_name, student_age, student_grade = line.strip().split(',')
                students[student_id] = {
                    'name': student_name,
                    'age': student_age,
                    'grade': student_grade
                }
    return students

def save_data(students):
    with open(FILE_NAME, 'w') as file:
        for student_id, student_info in students.items():
            file.write(f"{student_id},{student_info['name']},{student_info['age']},{student_info['grade']}\n")

def add_student(students, student_id, student_name, student_age, student_grade):
    if student_id in students:
        return "Student ID already exists. Please enter a unique ID."
    students[student_id] = {
        'name': student_name,
        'age': student_age,
        'grade': student_grade
    }
    save_data(students)
    return "Student added successfully!"

def search_student(students, student_id):
    if student_id in students:
        student = students[student_id]
        return f"ID: {student_id}, Name: {student['name']}, Age: {student['age']}, Grade: {student['grade']}"
    else:
        return "Student not found."

def delete_student(students, student_id):
    if student_id in students:
        del students[student_id]
        save_data(students)
        return "Student deleted successfully!"
    else:
        return "Student not found."

def list_students(students):
    if students:
        return [f"ID: {student_id}, Name: {student_info['name']}, Age: {student_info['age']}, Grade: {student_info['grade']}" for student_id, student_info in students.items()]
    else:
        return "No students found."

# Load Initial Data

students = load_data()
print("Initial data loaded:", students)

# Add a New Student

student_id = '004'
student_name = 'David'
student_age = '23'
student_grade = 'B'

result = add_student(students, student_id, student_name, student_age, student_grade)
print(result)

# Search for a Student

search_id = '002'
result = search_student(students, search_id)
print(result)

# Delete a Student

delete_id = '003'
result = delete_student(students, delete_id)
print(result)

# List All Students

result = list_students(students)
print(result)
```
